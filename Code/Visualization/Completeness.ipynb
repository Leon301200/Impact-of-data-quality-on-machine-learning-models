```python
import json
import matplotlib.pyplot as plt

# Charger le JSON depuis le fichier
with open('../../Results/FPS in video games/Completeness.json', 'r') as f:
    data = json.load(f)

# Extraire les informations nécessaires
models = data['models']

# Initialiser les métriques et les stratégies
metrics = ["mean_squared_error", "mean_absolute_error", "r2_score"]
metric_labels = {
    "mean_squared_error": "Erreur Quadratique Moyenne (MSE)",
    "mean_absolute_error": "Erreur Absolue Moyenne (MAE)",
    "r2_score": "Score R²"
}

strategies = ["Mean and Mode", "Median and New", "Decision Tree", "Mean and New"]

# Créer un graphique par stratégie
for strategy in strategies:
    fig, axs = plt.subplots(1, len(metrics), figsize=(15, 5))
    fig.suptitle(f"Stratégie d'imputation: {strategy}", fontsize=16)

    for col, metric in enumerate(metrics):
        ax = axs[col]

        for model in models:
            strategy_data = next(item for item in model['imputation_strategies'] if item['strategy'] == strategy)

            pollution_percentages = []
            metric_values = []

            for pollution_data in strategy_data['pollution_metrics']:
                pollution_percentages.append(pollution_data['pollution_percentage'])
                metric_values.append(pollution_data['metrics'][metric])

            ax.plot(pollution_percentages, metric_values, marker='o', label=model['model'])

        # Configuration des axes
        ax.set_title(metric_labels[metric])
        ax.set_xlabel("Pourcentage de Pollution")
        if col == 0:
            ax.set_ylabel("Valeur de la métrique")
        ax.grid(True)

    # Ajouter les légendes en dehors des sous-graphes
    fig.legend([model['model'] for model in models], loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=2)

    # Ajuster les espaces entre les sous-graphes
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
# Initialisation de la figure avec 4 sous-graphiques côte à côte
fig, axs = plt.subplots(1, 4, figsize=(20, 6), sharey=True)

# Liste des couleurs pour les différents modèles
colors = ['blue', 'green', 'red', 'orange']

# Noms des stratégies
strategies = ['Mean and Mode', 'Median and New', 'Mean and New', 'Decision Tree']

# Parcourir chaque stratégie
for j, strategy_name in enumerate(strategies):
    ax = axs[j]
    ax.set_title(strategy_name)

    # Parcourir chaque modèle
    for i, model_data in enumerate(data['models']):
        model_name = model_data['model']

        # Rechercher la stratégie correspondante dans les données
        strategy_data = next((strategy for strategy in model_data['imputation_strategies'] if strategy['strategy'] == strategy_name), None)

        if strategy_data:
            # Extraire les valeurs de pollution et de R² score
            pollution = [item['pollution_percentage'] for item in strategy_data['pollution_metrics']]
            r2_scores = [item['metrics']['r2_score'] for item in strategy_data['pollution_metrics']]

            # Tracer les courbes
            ax.plot(pollution, r2_scores, label=f'{model_name}', color=colors[i], marker='o')

    ax.set_xlabel('Pollution Percentage')
    if j == 0:
        ax.set_ylabel('R² Score')
    ax.grid(True)

# Ajouter la légende une seule fois pour tout le graphique
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=4, bbox_to_anchor=(0.5, 1.05))

plt.tight_layout()
plt.show()
```
