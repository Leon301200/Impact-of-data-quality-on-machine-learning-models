```python
import pandas as pd
```

```python
data = pd.read_csv('../../Data/Clustering/retail Data/retail_data.csv')
data
```

```python
# Supprimer les lignes avec au moins une valeur manquante
df = data.dropna()
df
```

```python
df = df.drop('Transaction_ID',axis=1)
df = df.drop('Customer_ID', axis=1)
df = df.drop('Name', axis=1)
df = df.drop('Email', axis=1)
df = df.drop('Phone', axis=1)
df = df.drop('Address', axis=1)
df = df.drop('Month', axis=1)
df = df.drop('Year', axis=1)
# df = df.drop('products', axis=1)
# df = df.drop('Product_Brand', axis=1)
df = df.drop('Time', axis=1)
df
```

```python
# Fonction pour convertir une date dans le bon format
def parse_date(date_str):
    formats = ['%d-%m-%y', '%m/%d/%Y']
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue
    return pd.NaT  # Si aucun format ne correspond

# df['date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df['Date'] = df['Date'].apply(parse_date)

# Extraire le jour, le mois et l'année
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

df = df.drop('Date', axis=1)
df[['Day','Month','Year']]
```
