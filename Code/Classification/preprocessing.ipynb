```python
# Import
import pandas as pd
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
```

```python
data = pd.read_csv('../../Data/Classification/Indicators of Heart Disease/heart_2022_no_nans.csv')
data
```

```python
df = data
pd.set_option('future.no_silent_downcasting', True)
df['HadHeartAttack'] = df['HadHeartAttack'].replace({'Yes': True, 'No': False}).astype(bool)
df['PhysicalActivities'] = df['PhysicalActivities'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadAngina'] = df['HadAngina'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadStroke'] = df['HadStroke'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadAsthma'] = df['HadAsthma'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadSkinCancer'] = df['HadSkinCancer'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadCOPD'] = df['HadCOPD'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadDepressiveDisorder'] = df['HadDepressiveDisorder'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadKidneyDisease'] = df['HadKidneyDisease'].replace({'Yes': True, 'No': False}).astype(bool)
df['HadArthritis'] = df['HadArthritis'].replace({'Yes': True, 'No': False}).astype(bool)
df['DeafOrHardOfHearing'] = df['DeafOrHardOfHearing'].replace({'Yes': True, 'No': False}).astype(bool)
df['BlindOrVisionDifficulty'] = df['BlindOrVisionDifficulty'].replace({'Yes': True, 'No': False}).astype(bool)
df['DifficultyConcentrating'] = df['DifficultyConcentrating'].replace({'Yes': True, 'No': False}).astype(bool)
df['DifficultyWalking'] = df['DifficultyWalking'].replace({'Yes': True, 'No': False}).astype(bool)
df['DifficultyDressingBathing'] = df['DifficultyDressingBathing'].replace({'Yes': True, 'No': False}).astype(bool)
df['DifficultyErrands'] = df['DifficultyErrands'].replace({'Yes': True, 'No': False}).astype(bool)
df['ChestScan'] = df['ChestScan'].replace({'Yes': True, 'No': False}).astype(bool)
df['AlcoholDrinkers'] = df['AlcoholDrinkers'].replace({'Yes': True, 'No': False}).astype(bool)
df['HIVTesting'] = df['HIVTesting'].replace({'Yes': True, 'No': False}).astype(bool)
df['FluVaxLast12'] = df['FluVaxLast12'].replace({'Yes': True, 'No': False}).astype(bool)
df['PneumoVaxEver'] = df['PneumoVaxEver'].replace({'Yes': True, 'No': False}).astype(bool)
df['HighRiskLastYear'] = df['HighRiskLastYear'].replace({'Yes': True, 'No': False}).astype(bool)
df
```

```python
df.to_csv('../../Data/Classification/Indicators of Heart Disease/heart_2022_no_nans_clean.csv', index=False)
```

```python
object_columns = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=object_columns)
df
```

```python
model = XGBClassifier()
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
X = df.drop('HadHeartAttack', axis=1)
y = df['HadHeartAttack']
results = cross_val_score(model, X, y, cv=kfold, scoring='f1')
```

```python
# Afficher les scores pour chaque pli
print(f"Scores pour chaque pli: {results}")

# Afficher la moyenne et l'écart-type des scores
print(f"Moyenne des scores: {results.mean()}")
print(f"Écart-type des scores: {results.std()}")
```
